```python
#Importamos librerias 
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import os
import seaborn as sns
%matplotlib inline
```

```python
#Utilizamos funcion para realizar la importacion del archivo
train = pd.read_csv("../USUARIO/TITANIC/TRAIN.CSV")
```

```python


#Con la siguiente funcion (info) obtendremos informacion de las columnas y los tipos de datos que contienen el conjunto de datos
train.info
```

```python
#Vemos las estadisticas basicas de las columnas con el uso de la funcion describe()
train.describe()
```

```python
#Utilizamos la funcion  "isna()"para conocer los datos no numericos y lurgo procedemos a sumarlos con "sum()"
train.isna().sum()
```

```python
#Como tenemoss valores numericos,en algunas variables, entonves procedemos a rellenaslos por la media de la misma variable, en este caso vamos a iniciar con la variable edad(Age) y al final los sumamos
media_edades=round(train["Age"].mean())
train["Age"]=train["Age"].fillna(media_edades)
train.isna().sum 
```

```python
#rellenamlos la variable cabina(cabin) por la media de los datos de la cabina, rellenamos con la abreviacion de no especificado, "NE"
train["Cabin"]=train["Cabin"].fillna("NE")
train.isna().sum()
```

```python
#Mostramos en pantalla los valores de cabina
train["Cabin"].value_counts()
```

```python
#Rellenamos los datos de l columna de embarque"Embarked", con la abreviacion "NE"
train["Embarked"]=train["Embarked"].fillna("NE")
train.isna().sum()
```

```python
#Contamos los valores 
train["Embarked"].value_counts()
```

```python
#Mostramos la lista de cabecera
train.head()
```

```python
train["Survived"]=train["Survived"].map({ 0:"No", 1: "Yes"})
train.head()
```

```python
#Cambiamos los datos de "S" por "Southampton", "C" por "Cherbourg" y "Q" por "Queenstowm" de la columna "Embarked"
train["Embarked"]=train["Embarked"].map({ "S": "Southampton","C":"Cherbourg", "Q":"Queenstowm"})
train.head
```

```python
#Una vez listo el conjunto de datos procedemos a ver informacion importante se puede obtener de él. Se creó una grafica de barras, en la que se visualiza el estado del pasajero, es decir si surió o sobrevivió, teniendo en cuenta la clase a la que pertenecía

train.groupby(["Pclass","Survived"])["Survived"].count()
ax=sns.countplot(x="Pclass", hue="Survived", palette="Set2", data=train)
ax.set(title="Estado del pasajero  (murió/sobrevivió) por la clase a la que pertenecía",
       xlabel="Clase del pasajero", ylabel="Total")
plt.show()
```

```python
train.groupby(["Sex", "Survived"])["Survived"].count()
```

```python
#Se crea una grafica del total de los pasajeros con respecto a su sexo

ax=sns.countplot(x="Sex", hue="Survived", palette="Set3", data=train)
ax.set(title="Total pasajeros con respecto a su sexo", xlabel="Sexo", ylabel="Total")
plt.show()
```

```python
ax=sns.catplot(x="Pclass", hue="Sex", col="Survived", palette="Set3", data=train, kind="count")
plt.show()
```
